# Explore the software architecture classification dataset

This notebook explores the software architecture classification dataset. It creates some graphic about distribution of the dataset.

```python
import os

data_dir = os.path.join('data', 'software_architecture')
bibtext_file = os.path.join(data_dir, 'bib-text.csv')
```

```python
import pandas as pd

df = pd.read_csv(bibtext_file)
```

## Research Objects

```python
from src.taxonomy.utils import parse_taxonomy

research_objects = []

for _, row in df.iterrows():
    taxonomy = parse_taxonomy(row['classes'])
    research_obj = taxonomy.children("Research Object")
    research_objects.append([obj.tag for obj in research_obj])
```

```python
# Plot the research fields distribution
import matplotlib.pyplot as plt

research_objects = [obj for sublist in research_objects for obj in sublist]
research_objects = pd.Series(research_objects)
research_objects.value_counts().plot(kind='bar')
plt.show()
```
