```python
import pandas as pd

chatgpt_evaluation = pd.read_csv('../../reports/openai_chatgpt/chatgpt_evaluation.csv')
```

```python
chatgpt_evaluation.head()
```

```python
# Plot the temperature as y axis and accuracy, precision, recall, f1-score as x axis
import matplotlib.pyplot as plt

plt.plot(chatgpt_evaluation['temperature'], chatgpt_evaluation['accuracy'], label='accuracy')
plt.plot(chatgpt_evaluation['temperature'], chatgpt_evaluation['precision'], label='precision')
plt.plot(chatgpt_evaluation['temperature'], chatgpt_evaluation['recall'], label='recall')
plt.plot(chatgpt_evaluation['temperature'], chatgpt_evaluation['f1'], label='f1')
plt.xlabel('temperature')
plt.ylabel('score')
plt.legend()
# Save the plot as png to the reports folder
plt.savefig('../../reports/openai_chatgpt/chatgpt_evaluation.png')

plt.show()
```
